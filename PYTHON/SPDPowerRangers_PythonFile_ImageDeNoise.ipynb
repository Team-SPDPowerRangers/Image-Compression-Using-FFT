FFt on images

```python
from matplotlib.image import imread
import numpy as np
import matplotlib.pyplot as plt
from scipy import io
```

```python
plt.rcParams['figure.figsize'] = [16,12]
plt.rcParams.update({'font.size':18})
```

```python
#A=io.imread()
A =imread("highimage1.jpg")
B =np.mean(A,-1)
```

```python
plt.figure()
plt.imshow(A)  #256-A
plt.axis('off')
```

```python
import matplotlib
from PIL import Image
import PIL
```

```python
Bt = np.fft.fft2(B)
Btsort = np.sort(np.abs(Bt.reshape(-1)))
#print(Btsort)
counter=1
for keep in (0.99,0.1,0.05,0.01,0.002):
    threshold = Btsort[int(np.floor((1-keep)*len(Btsort)))]
    #print(threshold)
    ind = np.abs(Bt)>threshold
    #print(ind)
    Btlow = Bt*ind
    #print(Btlow)
    Alow = np.fft.ifft2(Btlow).real    
    plt.figure()
    #Alow=np.array(Alow)
    #AAlow = Image.fromarray(Alow)
    #AAlow.save(str(counter)+"_highimage1.png")
    #Alow=Alow.save(str(counter)+"_highimage1.jpg")
    plt.imsave(Alow,str(counter)+"_highimage1.jpg")
    plt.imshow(Alow,cmap='gray')
    plt.axis('off')
    plt.title('Compressed image : keep = '+str(keep*100)+'%')
    counter=counter+1
```
